```python
import math

def calculate_entropy(pos, neg):
    total = pos + neg
    if total == 0:
        return 0
    return -((pos / total) * math.log2(pos / total) + (neg / total) * math.log2(neg / total))

def main():
    with open("exp4_input.csv", 'r') as file:
        lines = file.readlines()

    Name, PriceLessThan5L, Safe = "Name", "PriceLessThan5L", "Safe"
    parent = {"Yes": 0, "No": 0}
    child = {}

    i = 0
    child_column_choice = 0

    for line in lines:
        values = line.strip().split(',')
        if i == 0:
            i += 1
            print("Enter Child Column Number : ")
            child_column_choice = int(input())
            continue

        if child_column_choice == 1:
            child_name = values[0]
        elif child_column_choice == 2:
            child_name = values[1]
        else:
            child_name = values[1]

        parent[values[2]] += 1
        if child_name not in child:
            child[child_name] = {"Yes": 0, "No": 0}
        child[child_name][values[2]] += 1

    pos = parent["Yes"]
    neg = parent["No"]
    total = pos + neg
    parent_entropy = calculate_entropy(pos, neg)
    print("Parent Entropy:", parent_entropy)

    child_entropy = 0
    for val, counts in child.items():
        pR = counts["Yes"]
        nR = counts["No"]
        tR = pR + nR
        if tR != 0 and pR != 0 and nR != 0:
            child_entropy += -((pR + nR) / total) * ((pR / tR) * math.log2(pR / tR) + (nR / tR) * math.log2(nR / tR))

    print("Child Entropy * Their proportion:", child_entropy)
    print("Info gain:", parent_entropy - child_entropy)

if __name__ == "__main__":
    main()
```
