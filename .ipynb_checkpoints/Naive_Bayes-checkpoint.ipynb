```python
import csv
from collections import defaultdict
from functools import reduce

def main():
    filename = "exp13_input.csv"
    data = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
    parent = defaultdict(int)
    count = 0
    result_class = []

    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        titles = next(csv_reader)
        
        for line in csv_reader:
            n = len(line)
            parent[line[-1]] += 1

            for i in range(1, n - 1):
                data[titles[i]][line[i]][line[-1]] += 1

            count += 1

    for key in parent:
        result_class.append(key)

    output = [1] * len(result_class)

    for condition, sub_data in data.items():
        while True:
            try:
                user_input = input(f"Enter {condition} condition:\n")
                curr = sub_data[user_input]
                break
            except KeyError:
                print("No match. Try again.")

        for i in range(len(result_class)):
            print(f"{sub_data[user_input][result_class[i]]} / {parent[result_class[i]]}")
            val = sub_data[user_input][result_class[i]] / parent[result_class[i]]
            output[i] *= val
            print(output[i])

    for i in range(len(result_class)):
        output[i] *= parent[result_class[i]] / count

    total_sum = sum(output)
    print(f"Sum: {total_sum}")

    print("Output:")
    for i in range(len(result_class)):
        print(f"{result_class[i]} {output[i]}")

        percentage = (output[i] / total_sum) * 100
        print(f"Percentage: {percentage}")

if __name__ == "__main__":
    main()
```
