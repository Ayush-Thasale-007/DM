```python
import math

sub_classes = []
main_class = {}
dist_val = {"day": set(), "outlook": set(), "temp": set(), "humidity": set(), "wind": set()}
dist_val_count = {"day": 0, "outlook": 0, "temp": 0, "humidity": 0, "wind": 0}
val_count = {"day": {"Yes": 0, "No": 0}, "outlook": {"Yes": 0, "No": 0}, "temp": {"Yes": 0, "No": 0},
              "humidity": {"Yes": 0, "No": 0}, "wind": {"Yes": 0, "No": 0}}
max_gain = float('-inf')
root = "null"

with open("exp12_ip_gain.csv", 'r') as file:
    lines = file.readlines()
    main_class_name = "playGame"
    
    for j, line in enumerate(lines):
        if j == 0:
            columns = line.strip().split(',')
            sub_classes.extend(columns[:-1])
            continue
        
        values = line.strip().split(',')
        day, outlook, temp, humidity, wind, playGame = values
        sub_classes.extend([day, outlook, temp, humidity, wind])
        
        dist_val["day"].add(day)
        dist_val["outlook"].add(outlook)
        dist_val["temp"].add(temp)
        dist_val["humidity"].add(humidity)
        dist_val["wind"].add(wind)
        
        main_class[playGame] = main_class.get(playGame, 0) + 1
        dist_val_count["day"] += 1
        dist_val_count["outlook"] += 1
        dist_val_count["temp"] += 1
        dist_val_count["humidity"] += 1
        dist_val_count["wind"] += 1
        
        val_count["day"][playGame] += 1
        val_count["outlook"][playGame] += 1
        val_count["temp"][playGame] += 1
        val_count["humidity"][playGame] += 1
        val_count["wind"][playGame] += 1

totR = main_class["Yes"] + main_class["No"]
mainC_gain = -((main_class["Yes"] / totR) * math.log2(main_class["Yes"] / totR) +
               (main_class["No"] / totR) * math.log2(main_class["No"] / totR))
print(f"Main Class Gain: {mainC_gain}")

def calculate_gain(sub_class, mainC_gain):
    totR = main_class["Yes"] + main_class["No"]
    ent = 0
    
    for dv in dist_val[sub_class]:
        tR = dist_val_count[dv]
        pR, nR = val_count[dv]["Yes"], val_count[dv]["No"]
        
        if pR != 0:
            ent += - (tR / totR) * ((pR / tR) * math.log2(pR / tR))
        if nR != 0:
            ent += - (tR / totR) * ((nR / tR) * math.log2(nR / tR))
    
    print(f"InfoGain ({sub_class} | playGame): {ent}")
    fw.write(f"InfoGain ({sub_class} | playGame),{ent}\n")
    
    gain = mainC_gain - ent
    print(f"Gain ({sub_class} | playGame): {gain}\n")
    fw.write(f"Gain ({sub_class} | playGame),{gain}\n")
    
    global max_gain, root
    if gain > max_gain:
        max_gain = gain
        root = sub_class

for i in range(1, 5):
    calculate_gain(sub_classes[i], mainC_gain)

print(f"Subclass: {root} has maximum gain. Hence it will be selected as the root for splitting.")
fw.write(f"Subclass: {root} has maximum gain. Hence it will be selected as the root for splitting.\n")

fw.close()
```

```python
def calculate_gini(sub_class, data):
    total_records = len(data)
    gini_index = 1.0
    unique_values = set(row[sub_class] for row in data)

    for value in unique_values:
        probability = sum(1 for row in data if row[sub_class] == value) / total_records
        gini_index -= probability ** 2

    return gini_index

def main():
    data = []
    with open("gini_ip.csv", 'r') as file:
        lines = file.readlines()
        main_class = "Decision"

        for j, line in enumerate(lines):
            if j == 0:
                continue

            values = line.strip().split(',')
            weekend, weather, parents, money, decision = values
            row = {"Weekend": weekend, "Weather": weather, "Parents": parents, "Money": money, "Decision": decision}
            data.append(row)

    gini_main_class = calculate_gini(main_class, data)
    print(f"Main Class Gini Index: {gini_main_class}")

    max_gini = 0
    subclass_max = ""
    for sub_class in data[0]:
        if sub_class != main_class:
            gini_sub_class = calculate_gini(sub_class, data)
            print(f"Gini Index for {sub_class}: {gini_sub_class}")

            if gini_sub_class > max_gini:
                max_gini = gini_sub_class
                subclass_max = sub_class

    print(f"Subclass: {subclass_max} has maximum gini index. Hence it will be selected as root for splitting")

if __name__ == "__main__":
    main()
```
