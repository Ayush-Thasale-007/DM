```python
from itertools import combinations

def wordsof(s):
    words = []
    tmp = ""
    i = 0
    while i < len(s):
        if s[i].isalnum():
            tmp += s[i]
        else:
            if tmp:
                words.append(tmp)
            tmp = ""
        i += 1
    if tmp:
        words.append(tmp)
    return words

def combine(arr, miss):
    return " ".join([arr[i] for i in range(len(arr)) if i != miss])

def cloneit(arr):
    return set(arr)

def apriori_gen(sets, k):
    set2 = set()
    for it1 in sets:
        for it2 in sets:
            v1 = wordsof(it1)
            v2 = wordsof(it2)
            alleq = all(v1[i] == v2[i] for i in range(k - 1))
            v1.append(v2[k - 1])
            v1[-2], v1[-1] = min(v1[-2], v1[-1]), max(v1[-2], v1[-1])
            alleq = alleq and all(v1[i] in sets for i in range(len(v1)))
            if alleq:
                set2.add(combine(v1, -1))
    return set2

def count_occurrences(v, datatable):
    count = 0
    for s in datatable:
        present = all(x in s for x in v)
        if present:
            count += 1
    return count

def subsets(items, v1, v2, idx, datatable):
    if idx == len(items):
        if not v1 or not v2:
            return
        count1 = count_occurrences(items, datatable)  # Total support
        count2 = count_occurrences(v1, datatable)
        conf = (count1 / count2) * 100 if count2 != 0 else 0
        if conf >= confidence:
            fw1.write(f"{{ {' '.join(v1)} }} -> {{ {' '.join(v2)} }} , {conf:.2f}\n")
        return

    v1.append(items[idx])
    subsets(items, v1, v2, idx + 1, datatable)
    v1.pop()

    v2.append(items[idx])
    subsets(items, v1, v2, idx + 1, datatable)
    v2.pop()

def generate_association_rules(freq_items, datatable):
    for item in freq_items:
        items = wordsof(item)
        subsets(items, [], [], 0, datatable)

if __name__ == "__main__":
    fin = open("exp8.csv", "r")
    lines = fin.readlines()

    datatable = []
    products = set()
    freq = {}
    minfre = 0
    confidence = 0
    fw1 = open("exp8_output.csv", "w")

    i = 0
    for line in lines:
        arr = wordsof(line.strip())
        tmpset = set(arr)
        datatable.append(tmpset)
        for item in tmpset:
            products.add(item)
            freq[item] = freq.get(item, 0) + 1

    fin.close()

    print("No of transactions:", len(datatable))
    minfre = float(input("Enter Support count: "))
    minfre = minfre * len(datatable) / 100
    print("Min frequency:", minfre)
    
    q = [item for item, count in freq.items() if count < minfre]
    for item in q:
        products.remove(item)

    pass_value = 1
    print(f"\nFrequent {pass_value}-item set :")
    for item in products:
        print(f"{{ {item} }} {freq[item]}")

    i = 2
    prev = cloneit(products)
    while i:
        cur = apriori_gen(prev, i - 1)
        if not cur:
            break
        for item in cur:
            arr = wordsof(item)
            tot = sum(1 for s in datatable if all(x in s for x in arr))
            if tot >= minfre:
                freq[item] += tot
            else:
                q.append(item)

        for item in q:
            cur.remove(item)

        flag = all(freq[item] >= minfre for item in cur)
        if not cur:
            break

        print(f"\n\nFrequent {pass_value + i - 1}-item set :")
        for item in cur:
            print(f"{{ {item} }} {freq[item]}")

        prev = cloneit(cur)
        i += 1

    generate_association_rules(prev, datatable)
    fw1.close()
```
