```python
import math

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def main():
    points = []
    with open("cluster_input.csv", 'r') as infile:
        next(infile)  # skip the header
        for line in infile:
            _, x, y, _ = line.strip().split(',')
            points.append((int(x), int(y)))

    n = len(points)
    x_sum = sum(x for x, _ in points)
    y_sum = sum(y for _, y in points)
    mid_x = x_sum / n
    mid_y = y_sum / n
    print(f"Mid Point: ({mid_x},{mid_y})")

    with open("cluster_output.csv", 'w') as outfile:
        outfile.write(" , p1 ,p2 ,p3 ,p4,C\n")
        for i in range(n):
            if i < n:
                outfile.write("p" + str(i + 1) + ",")
            for j in range(i + 1):
                x1, y1 = points[i]
                x2, y2 = points[j]
                if x1 == x2 and y1 == y2:
                    outfile.write("0,")
                    break
                dis = distance(x1, y1, x2, y2)
                outfile.write(str(dis) + ",")
            outfile.write("\n")

        outfile.write("C,")
        nearest_point = min(points, key=lambda p: distance(mid_x, mid_y, p[0], p[1]))
        ans = points.index(nearest_point) + 1
        x_new, y_new = nearest_point
        nearer = float('inf')
        for i, (x, y) in enumerate(points):
            d = distance(mid_x, mid_y, x, y)
            print(f"Distance of p{i + 1} from centre: {d}")
            outfile.write(str(d) + ",")
            if nearer > d:
                nearer = d
                ans = i + 1
                x_new, y_new = x, y
            if i == n - 1:
                outfile.write("0,")

        print(f"Nearer Distance: {nearer}")
        print(f"Nearest point from Centre is: p{ans}")

        outfile.write("\n,")
        outfile.write("\n")
        outfile.write(" , p1 ,p2 ,p3 ,p4\n")
        for i in range(n):
            if i < n:
                outfile.write("p" + str(i + 1) + ",")
            for j in range(i + 1):
                x1, y1 = points[i]
                x2, y2 = points[j]
                if x1 == x2 and y1 == y2:
                    outfile.write("0,")
                    break
                dis = distance(x1, y1, x2, y2)
                outfile.write(str(dis) + ",")
            outfile.write("\n")

        outfile.write("p" + str(ans) + " (New Center),")
        for i, (x, y) in enumerate(points):
            d = distance(x_new, y_new, x, y)
            print(f"Distance of p{i + 1} from p{ans}: {d}")
            outfile.write(str(d) + ",")
            if i == n - 1:
                outfile.write("0,")

if __name__ == "__main__":
    main()
```
