```python
import csv
import matplotlib.pyplot as plt
```

```python
data = []
with open("abalone.csv", "r") as file:
    reader = csv.reader(file)
    data.append(list(reader))
```

```python
feature = []
for val in data[0]:
    feature.append(float(val[1]))
```

```python
min_val = min(feature)
max_val = max(feature)
```

```python
normalized = []
```

```python
min_ = 1
max_ = 10

for val in feature:
    val = float(val)
    min_val, max_val = float(min_val), float(max_val)
    deno = (val - min_val)/(max_val - min_val)
    mul = max_ - min_
    normalized.append((deno*mul)+min_)
```

```python
min(normalized)
```

```python
var = 0
mean_ = 0
n = len(feature)
sum = 0
for val in feature:
    sum+=val

mean_ = sum/n    

for val in feature:
    var += (val-mean_)**2 

var = var/n
std_dev = var**(1/2)
```

```python
normalized = []

for val in feature:
    normalized.append((val-mean_)/std_dev)
```
