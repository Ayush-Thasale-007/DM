```python
import csv

def wordsof(s):
    words = []
    tmp = ""
    for char in s:
        if char.isalnum():
            tmp += char
        else:
            if tmp:
                words.append(tmp)
            tmp = ""
    if tmp:
        words.append(tmp)
    return words

def combine(arr, miss):
    return " ".join(word for i, word in enumerate(arr) if i != miss)

def cloneit(arr):
    return set(arr)

def apriori_gen(sets, k):
    set2 = set()
    for i in range(len(sets)):
        for j in range(i + 1, len(sets)):
            v1 = wordsof(sets[i])
            v2 = wordsof(sets[j])
            alleq = all(v1[l] == v2[l] for l in range(k - 1))
            v1.append(v2[k - 1])
            if v1[-1] < v1[-2]:
                v1[-1], v1[-2] = v1[-2], v1[-1]
            alleq = alleq and all(combine(v1, l) in sets for l in range(len(v1)))
            if alleq:
                set2.add(combine(v1, -1))
    return set2

minfre = float(input("Frequency % :"))
datatable = []
products = set()
freq = {}

with open("apriori.csv", "r") as fin:
    reader = csv.reader(fin)
    for row in reader:
        arr = wordsof(row[0])
        tmpset = set(arr)
        datatable.append(tmpset)
        for product in tmpset:
            products.add(product)
            freq[product] = freq.get(product, 0) + 1

print("No of transactions:", len(datatable))
minfre = minfre * len(datatable) / 100
print("Min frequency:", minfre)

q = [product for product in products if freq[product] < minfre]
for product in q:
    products.remove(product)

pass_num = 1
print("\nFrequent", pass_num, "-item set :\n")
for product in products:
    print("{" + product + "}", freq[product])

i = 2
prev = cloneit(products)
while i:
    cur = apriori_gen(prev, i - 1)
    if len(cur) < 1:
        break
    for item in cur:
        arr = wordsof(item)
        tot = sum(1 for transaction in datatable if all(product in transaction for product in arr))
        if tot >= minfre:
            freq[item] = freq.get(item, 0) + tot
        else:
            q.append(item)

    while q:
        cur.remove(q.pop(0))

    flag = all(freq[item] >= minfre for item in cur)

    if not cur:
        break

    print("\n\nFrequent", pass_num, "-item set :\n")
    for item in cur:
        print("{" + item + "}", freq[item])

    prev = cloneit(cur)
    i += 1

with open("output.csv", "w", newline='') as fw:
    writer = csv.writer(fw)
    for item in prev:
        writer.writerow(["{" + item + "}"])
```
